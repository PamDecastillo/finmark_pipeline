```python
import pandas as pd
import os

print("Libraries imported successfully ✅")
```

```python
trend_df = pd.read_csv("data/raw/trend_report.csv")
marketing_df = pd.read_csv("data/raw/marketing_summary.csv")
events_df = pd.read_csv("data/raw/event_logs.csv")

print("trend_report shape:", trend_df.shape)
print("marketing_summary shape:", marketing_df.shape)
print("event_logs shape:", events_df.shape)
```

```python
print("=== TREND REPORT ===")
display(trend_df.head())

print("=== MARKETING SUMMARY ===")
display(marketing_df.head())

print("=== EVENT LOGS ===")
display(events_df.head())
```

```python
print("=== Nulls in Trend Report ===")
print(trend_df.isnull().sum())

print("\n=== Nulls in Marketing Summary ===")
print(marketing_df.isnull().sum())

print("\n=== Nulls in Event Logs ===")
print(events_df.isnull().sum())
```

```python
# Keep only useful columns
trend_clean = trend_df[["week", "avg_users", "sales_growth_rate"]].copy()

# Parse date
trend_clean["week"] = pd.to_datetime(
    trend_clean["week"] + "-1", format="%Y-W%W-%w", errors="coerce"
)

# Fill nulls with median
trend_clean["avg_users"] = trend_clean["avg_users"].fillna(trend_clean["avg_users"].median())
trend_clean["sales_growth_rate"] = trend_clean["sales_growth_rate"].fillna(0)

print("Trend Report Cleaned ✅")
display(trend_clean.head())
```

```python
# Keep only useful columns
marketing_clean = marketing_df[["date", "users_active", "total_sales", "new_customers"]].copy()

# Parse date
marketing_clean["date"] = pd.to_datetime(marketing_clean["date"], errors="coerce")

# Fill nulls
marketing_clean["total_sales"] = marketing_clean["total_sales"].fillna(marketing_clean["total_sales"].median())
marketing_clean["users_active"] = marketing_clean["users_active"].fillna(0)
marketing_clean["new_customers"] = marketing_clean["new_customers"].fillna(0)

print("Marketing Summary Cleaned ✅")
display(marketing_clean.head())
```

```python
import hashlib

# Keep only useful columns
events_clean = events_df[["user_id", "event_type", "event_time", "product_id"]].copy()

# Parse datetime
events_clean["event_time"] = pd.to_datetime(events_clean["event_time"], errors="coerce")

# Remove duplicates
events_clean = events_clean.drop_duplicates(subset=["user_id", "event_type", "event_time"])

# Keep only valid event types
valid_events = ["checkout", "page_view", "wishlist_add", "add_to_cart", "login", "search", "profile_update"]
events_clean = events_clean[events_clean["event_type"].isin(valid_events)]

# Mask user IDs
events_clean["user_id_masked"] = events_clean["user_id"].apply(
    lambda uid: "U_" + hashlib.sha256(str(uid).encode()).hexdigest()[:8]
)

print("Event Logs Cleaned ✅")
print("Total rows:", len(events_clean))
display(events_clean.head())
```

```python
marketing_kpi = marketing_clean.copy()
marketing_kpi = marketing_kpi.sort_values("date")

# Sales per user
marketing_kpi["sales_per_user"] = (marketing_kpi["total_sales"] / marketing_kpi["users_active"]).round(2)

# 7 day rolling average
marketing_kpi["sales_7d_avg"] = marketing_kpi["total_sales"].rolling(window=7, min_periods=1).mean().round(2)

# Below trend flag
marketing_kpi["below_trend"] = marketing_kpi["total_sales"] < marketing_kpi["sales_7d_avg"]

print("KPIs Added ✅")
display(marketing_kpi.head(10))
```

```python
events_clean["event_date"] = events_clean["event_time"].dt.date

funnel = events_clean.groupby(["event_date", "event_type"]).size().unstack(fill_value=0).reset_index()

if "page_view" not in funnel.columns:
    funnel["page_view"] = 0
if "checkout" not in funnel.columns:
    funnel["checkout"] = 0

funnel["conversion_rate"] = (funnel["checkout"] / funnel["page_view"].replace(0, 1)).round(4)

print("Funnel Built ✅")
display(funnel)
```

```python
trend_processed = trend_clean.copy()

trend_processed["growth_flag"] = trend_processed["sales_growth_rate"].apply(
    lambda x: "Rising" if x > 0.05 else ("Falling" if x < -0.05 else "Stable")
)

print("Growth Flags Added ✅")
display(trend_processed)
```

```python
os.makedirs("data/cleaned", exist_ok=True)
os.makedirs("data/processed", exist_ok=True)

trend_clean.to_csv("data/cleaned/trend_report_cleaned.csv", index=False)
marketing_clean.to_csv("data/cleaned/marketing_summary_cleaned.csv", index=False)
events_clean.to_csv("data/cleaned/event_logs_cleaned.csv", index=False)

marketing_kpi.to_csv("data/processed/marketing_kpis.csv", index=False)
funnel.to_csv("data/processed/event_funnel_daily.csv", index=False)
trend_processed.to_csv("data/processed/trend_report_processed.csv", index=False)

print("All files saved! ✅")
print("Cleaned → data/cleaned/")
print("Processed → data/processed/")
```
